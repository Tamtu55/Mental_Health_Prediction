```python
import numpy as np
import pandas as pd

# Define the number of rows in the dataset
num_rows = 1000

# Generate random data for each column
gender = np.random.randint(0, 2, size=num_rows)  # Binary gender values (0 or 1)
age = np.random.randint(18, 80, size=num_rows)  # Random age between 18 and 80
screentime = np.random.randint(0, 24, size=num_rows)  # Random screentime hours per day
most_used_app = np.random.randint(0, 2, size=num_rows)  # Binary most used app values (0 or 1)
emotional_support = np.random.randint(1, 11, size=num_rows)  # Random emotional support rating
self_care = np.random.randint(0, 2, size=num_rows)  # Binary self care values (0 or 1)
depression = np.random.randint(0, 2, size=num_rows)  # Binary depression values (0 or 1)
anxiety = np.random.randint(0, 2, size=num_rows)  # Binary anxiety values (0 or 1)
panic_attack = np.random.randint(0, 2, size=num_rows)  # Binary panic attack values (0 or 1)

# Create a DataFrame from the generated data
data = pd.DataFrame({
    'Gender': gender,
    'Age': age,
    'Screentime': screentime,
    'Most used App': most_used_app,
    'Emotional Support': emotional_support,
    'Self Care': self_care,
    'Depression': depression,
    'Anxiety': anxiety,
    'Panic_Attack': panic_attack
})

# Save the dataset to a CSV file
data.to_csv('synthetic_dataset.csv', index=False)

```

```python
x = data[['Gender', 'Age', 'Screentime', 'Most used App', 'Emotional Support', 'Self Care', 'Anxiety', 'Panic_Attack']]
Y = data['Depression']
```

```python
x.shape,Y.shape
```

```python
 ### Train Test Split
from sklearn.model_selection import train_test_split
x_train,x_test,Y_train,Y_test=train_test_split(x,Y,test_size=0.3,random_state=0)
```

```python
## Implement Random Forest classifier
from sklearn.ensemble import RandomForestClassifier
synthetic=RandomForestClassifier()
synthetic.fit(x_train,Y_train)
```

```python
## Prediction
Y_pred=synthetic.predict(x_test)
```

```python
## Check Accuracy
from sklearn.metrics import accuracy_score
score=accuracy_score(Y_test,Y_pred)
```
